```python
import nltk
from nltk import bigrams
from nltk import trigrams
from nltk.tokenize import word_tokenize
```

```python
#job_descriptions['tokens'] = nltk.word_tokenize(job_descriptions['FullDescription'])
#nltk.download()
```

```python
job_desc_series = job_descriptions["FullDescription"]
```

```python
descript_string = " ".join(job_desc_series)
```

```python
len(descript_string)
```

```python
descript_string = descript_string.decode("utf-8")
```

```python
words = word_tokenize(descript_string)
#fdist = nltk.FreqDist(words)
```

```python
words
```

```python
fdist = nltk.FreqDist(words)
```

```python
word_list = []
frequency_list = []
i = 1
for word, frequency in fdist.most_common(100):
    word_list.append(i)
    frequency_list.append(frequency)
    i+= 1
    print('%s;%d' % (word, frequency)).encode('utf-8')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot(word_list, frequency_list)
plt.title("Zipf's Law")
plt.xlabel('Word Rank')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.savefig('Zipfs_Law')
```

```python
from nltk.corpus import stopwords
```

```python
default_stopwords = set(nltk.corpus.stopwords.words('english'))
words = [word.lower() for word in words]
```

```python
words = [word for word in words if word not in default_stopwords]
words = [word for word in words if len(word) > 1]
words = [word for word in words if word != "****"]
```

```python
fdist2 = nltk.FreqDist(words)
for word, frequency in fdist2.most_common(10):
    print('%s;%d' % (word, frequency)).encode('utf-8')
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
lem_words = [wordnet_lemmatizer.lemmatize(word) for word in words]
```

```python
len(lem_words)
```

```python
fdist2 = nltk.FreqDist(lem_words)
for word, frequency in fdist2.most_common(10):
    print('%s;%d' % (word, frequency)).encode('utf-8')
```
